```python
# Import library used to work with tabular data like spreadsheets, CSVs, and SQL tables
import pandas as pd
```

```python
# Describing a list of feature names
features = [
    "Fwd Seg Size Min",
    "Init Bwd Win Byts",
    "Init Fwd Win Byts",
    "Fwd Seg Size Min",
    "Fwd Pkt Len Mean",
    "Fwd Seg Size Avg",
    "Label",
    "Timestamp",
]
```

```python
# Defining types of features
dtypes = {
    "Fwd Pkt Len Mean": "float",
    "Fwd Seg Size Avg": "float",
    "Init Fwd Win Byts": "int",
    "Init Bwd Win Byts": "int",
    "Fwd Seg Size Min": "int",
    "Label": "str",
}
```

```python
# Loading the dataset into a pandas DataFrame
df = pd.read_csv(
    "ddos_dataset.csv",
    usecols=features,
    dtype=dtypes,
    index_col=None,
)
```

```python
# Parse date column
import numpy as np

def try_parse(timestamp):
    try:
        return pd.to_datetime(timestamp, format="%d/%m/%Y %I:%M:%S %p")
    except:
        try:
            return pd.to_datetime(timestamp, format="%d/%m/%Y %H:%M:%S")
        except:
            return np.nan

df["Timestamp"] = df["Timestamp"].apply(try_parse)
```

```python
# Sorting the dataset chronologically
df2 = df.sort_values("Timestamp")
```

```python
# Removing the non-numeric, non-feature column: "Timestamp"
df3 = df2.drop(columns=["Timestamp"])
```

```python
# Split Into Training and Testing Sets
l = len(df3.index)
train_df = df3.head(int(l * 0.8))
test_df = df3.tail(int(l * 0.2))
```

```python
# Check Class Distribution
from collections import Counter

print(Counter(train_df["Label"].values))
print(Counter(test_df["Label"].values))
```

```python
# Separate Features and Labels
y_train = train_df.pop("Label").values
y_test = test_df.pop("Label").values
```

```python
# Convert DataFrames to NumPy Arrays
X_train = train_df.values
X_test = test_df.values
```

```python
# Initializing Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=50)
```

```python
# Training the Random Forest Classifie
clf.fit(X_train, y_train)
```

```python
# Evaluate Training Accuracy
clf.score(X_train, y_train)
```

```python
# Evaluate Test Accuracy
clf.score(X_test, y_test)
```

```python
# Plot the confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Predict on test set
y_pred = clf.predict(X_test)

# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```

```python
# Classification report
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
# ROC Curve  visualizes the true positive rate vs. false positive rate.

from sklearn.preprocessing import LabelEncoder

# Create and fit a label encoder
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
y_pred = clf.predict(X_test)

from sklearn.metrics import roc_curve, auc

y_proba = clf.predict_proba(X_test)[:, 1]  # Probability for class 1
fpr, tpr, _ = roc_curve(y_test, y_proba)
roc_auc = auc(fpr, tpr)

# Plot
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, label=f"ROC curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], "k--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Feature importance
import pandas as pd

feat_importances = pd.Series(clf.feature_importances_, index=train_df.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.title("Top 10 Feature Importances")
plt.xlabel("Importance")
plt.tight_layout()
plt.show()
```

🧠 Interpretation:
The model is very cautious when predicting DDoS, leading to zero false positives.
However, it's missing half of the actual DDoS attacks, which is critical in security contexts.
This suggests a recall-precision tradeoff — great precision, but recall needs improvement.
