```python
# Import required libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  

# Load dataset
df = pd.read_csv('datasets/sms_spam_svm.csv')
df.head()
```

```python
# Extract type and map it to 1 or -1 values
y = df.iloc[:, 0].values
y = np.where(y == 'spam', -1, 1)
# Extracting features: suspect and neutral
X = df.iloc[:, [1, 2]].values
```

```python
from sklearn.model_selection import train_test_split

# Splitting the data into train and test 70/30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.svm import SVC

# Creating and training the SVM Model with a linear kernel 
svm = SVC(kernel='linear', C=1.0, random_state=1)
svm.fit(X_train, y_train)

# Make predicting
y_pred = svm.predict(X_test)
```

```python
from defs import plot_decision_regions

# Combining train and test data
X_combined = np.vstack((X_train, X_test))
y_combined = np.hstack((y_train, y_test))

# Plotting the Decision Regions
plot_decision_regions(X_combined, y_combined, classifier=svm, test_idx=range(-15, 15))
plt.xlabel('suspect words')
plt.ylabel('spam or ham')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import accuracy_score

# evaluating the performance of your trained SVM model
print('Misclassified samples: %d' % (y_test != y_pred).sum())
print('Accuracy: %.2f' % accuracy_score(y_test, y_pred))
```

84% accuracy is a solid start, especially with just two features and a simple linear SVM
However, 7 misclassifications could be significant depending on the total number of test samples and the application (e.g., misclassifying spam as ham can be annoying, but misclassifying ham as spam could be critical).

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Compute the confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=[1, -1])

# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Ham', 'Spam'])
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()
```
