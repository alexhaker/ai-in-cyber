```python
import warnings
warnings.simplefilter('ignore')

# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('datasets/sms_spam_perceptron.csv')
df.head()
```

```python
# Map target type to -1 and 1 accordingly
y = df.iloc[:, 0].values
y = np.where(y == 'spam', -1, 1)

# Extract features: sex and buy
X = df.iloc[:, [1, 2]].values
```

```python
from sklearn.model_selection import train_test_split

# Divide the dataset into the train and test data 70/30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import Perceptron

# Training a Perceptron model
p = Perceptron(max_iter=40, eta0=0.1, random_state=0)
p.fit(X_train, y_train)
```

```python
# Run predictions on test data
y_pred = p.predict(X_test)
```

```python
from defs import plot_decision_regions

# Merging train and test sets
X_combined = np.vstack((X_train, X_test))
y_combined = np.hstack((y_train, y_test))

# Plotting the Decision Regions
plot_decision_regions(X=X_combined, y=y_combined, classifier=p, test_idx=range(-5, 5))
plt.xlabel('suspect words')
plt.ylabel('spam or ham')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import accuracy_score

# evaluating Perceptron model’s performance
print('Misclassified samples: %d' % (y_test != y_pred).sum())
print('Accuracy: %.2f' % accuracy_score(y_test, y_pred))
```

For a simple linear model using only two features, this is a decent result
20% error rate is relatively high for spam detection tasks, which often require >95% accuracy in production.
The two features may not be enough to capture the full context 

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Compute the confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=[1, -1])  # 1 = ham, -1 = spam

# Create a display object
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Ham', 'Spam'])

# Plot
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
# Let's add weights to the features
# sex - 0.5x, buy - 2x
feature_weights = np.array([0.5, 3])
X_weighted = X * feature_weights

X_train_w, X_test_w, y_train, y_test = train_test_split(X_weighted, y, test_size=0.3, random_state=0)

p_weighted = Perceptron(max_iter=40, eta0=0.1, random_state=0)
p_weighted.fit(X_train_w, y_train)
y_pred_w = p_weighted.predict(X_test_w)
```

```python
# Visualise results of prediction
X_combined_w = np.vstack((X_train_w, X_test_w))
y_combined = np.hstack((y_train, y_test))

plot_decision_regions(X=X_combined_w, y=y_combined,
                      classifier=p_weighted, test_idx=range(len(X_train_w), len(X_combined_w)))
plt.xlabel('Sex (Weighted)')
plt.ylabel('Buy (Weighted)')
plt.title('Perceptron Decision Boundary with Feature Weights')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
# Evaluate model
print('Misclassified samples: %d' % (y_test != y_pred_w).sum())
print('Accuracy: %.2f' % accuracy_score(y_test, y_pred_w))
```

```python
# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred_w, labels=[1, -1])  # 1 = ham, -1 = spam

# Create and display matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Ham', 'Spam'])
disp.plot(cmap=plt.cm.Blues)

plt.title('Confusion Matrix (Weighted Features)')
plt.grid(False)
plt.tight_layout()
plt.show()
```
